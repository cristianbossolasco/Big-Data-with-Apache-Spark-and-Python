```python
import pandas as pd
```

```python
data = [
    "SOLUSDT Perpetuo Cuadrícula de futuros",
    "501.75187192 USDT",
    "0.01827735",
    "≈ ARS$488,654.33",
    "1.74978961 USDT",
    "Crear",
    "Ver",
    "BTCUSDT Perpetuo Cuadrícula de futuros",
    "500.71287229 USDT",
    "0.01823950",
    "≈ ARS$487,641.47",
    "0.71860504 USDT",
    "Crear",
    "Ver",
    "WLDUSDT Perpetuo Cuadrícula de futuros",
    "1054.08791697 USDT",
    "0.03839734",
    "≈ ARS$1,026,578.25",
    "2.67053600 USDT",
    "Crear",
    "Ver",
    "LINKUSDT Perpetuo Cuadrícula de futuros",
    "501.39775288 USDT",
    "0.01826445",
    "≈ ARS$488,313.46",
    "1.40054403 USDT",
    "Crear",
    "Ver"
]

num_columns = 7  # Número de columnas en la tabla
dias = 1
```

```python
# Dividir los datos en filas
table = [data[i:i+num_columns] for i in range(0, len(data), num_columns)]

# Crear un DataFrame de Pandas
df = pd.DataFrame(table, columns=["Estrategia", "Importe", "ValorEnBTC", "ValorEnARS", "Ganancias", "Accion1", "Accion2"])

df["Importe"] = df["Importe"].str.replace(' USDT', '').astype(float)
df["Ganancias"] = df["Ganancias"].str.replace(' USDT', '').astype(float)
df["ImporteOriginal"] = df["Importe"] - df["Ganancias"]
df["PorcGanado"] = df["Ganancias"] / df["ImporteOriginal"] * 100
df["PorcAnual"] = df["PorcGanado"]  / dias * 365


SumaGananciasTotales = df["Ganancias"].sum()
SumaImporteOriginal = df["ImporteOriginal"].sum()
PorcGanado = SumaGananciasTotales / SumaImporteOriginal * 100
PorcGanadoAnual = PorcGanado / dias * 365
print(f'Porcentaje ganado = {round(PorcGanado, 2)} %')
print(f'Ganancias anuales = {round(PorcGanadoAnual, 2)} %')
```
